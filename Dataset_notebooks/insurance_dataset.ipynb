```python
over = [] 
for i in range(len(insurance)):
    if insurance.iloc[i]['bmi'] > 25:
        over.append(1)
    else: 
        over.append(0)
```

```python
insurance['overweight'] = over
```

```python
ohe = DataHelper().OHE(insurance['region'])
insurance2 = pd.concat([insurance, ohe], axis='columns')
insurance2
```

```python
sex = []
smoker = []
for i in range(len(insurance)):
    if insurance.iloc[i]['sex'] == 'female':
        sex.append(0)
    else:
        sex.append(1)
    if insurance.iloc[i]['smoker'] == 'yes':
        smoker.append(1)
    else:
        smoker.append(0)
```

```python
insurance2['sex_int'] = sex
insurance2['smoker'] = smoker
```
